```python
import pandas as pd
```

```python
def merge_df(nft_label):
    df = pd.merge(pd.read_csv('./data/{}/1.csv'.format(nft_label)), pd.read_csv('./data/{}/2.csv'.format(nft_label)))
    df = pd.merge(df, pd.read_csv('./data/{}/3.csv'.format(nft_label)))
    df['DateTime'] = pd.to_datetime(df['DateTime'], format = '%Y-%m-%d')
    return df
```

```python
df_cryptopunk = merge_df('cryptopunk')
df_artblocks = merge_df('artblocks')
df_bayc = merge_df('bayc')

df_cryptopunk = df_cryptopunk[df_cryptopunk['DateTime']>'2022-05-15']
df_artblocks = df_artblocks[df_artblocks['DateTime']>'2022-05-15']
df_bayc = df_bayc[df_bayc['DateTime']>'2022-05-15']

prev_df_cryptopunk = pd.read_csv('./data/prev_cryptopunk.csv')[1:]
prev_df_artblocks = pd.read_csv('./data/prev_artblocks.csv')
prev_df_bayc = pd.read_csv('./data/prev_bayc.csv')

prev_df_cryptopunk['DateTime'] = pd.to_datetime(prev_df_cryptopunk['DateTime'], format = '%Y-%m-%d')
prev_df_artblocks['DateTime'] = pd.to_datetime(prev_df_artblocks['DateTime'], format = '%Y-%m-%d')
prev_df_bayc['DateTime'] = pd.to_datetime(prev_df_bayc['DateTime'], format = '%Y-%m-%d')

prev_df_cryptopunk = prev_df_cryptopunk[list(df_cryptopunk)]
prev_df_artblocks = prev_df_artblocks[list(df_artblocks)]
prev_df_bayc = prev_df_bayc[list(df_bayc)]
```

```python
len(df_cryptopunk)+len(prev_df_cryptopunk)
```

```python
def combine_prev_and_lastest(df, prev_df):
    df = pd.concat([prev_df, df], ignore_index=True)
    df = df.dropna()
    df = df[df['Average USD'] != 0]
    return df
```

```python
final_df_cryptopunk = combine_prev_and_lastest(df_cryptopunk, prev_df_cryptopunk).reset_index(drop=True)
final_df_artblocks = combine_prev_and_lastest(df_artblocks, prev_df_artblocks).reset_index(drop=True)
final_df_bayc = combine_prev_and_lastest(df_bayc, prev_df_bayc).reset_index(drop=True)
```
